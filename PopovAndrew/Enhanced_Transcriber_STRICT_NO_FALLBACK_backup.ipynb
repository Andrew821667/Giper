```python
# üö® –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –£–°–¢–ê–ù–û–í–ö–ê –õ–û–ö–ê–õ–¨–ù–´–• –ó–ê–í–ò–°–ò–ú–û–°–¢–ï–ô
import os
import sys
import subprocess
from pathlib import Path
import time

# –°–¢–†–û–ì–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ä–µ–¥—ã
try:
    import google.colab
    IN_COLAB = True
    print("üéØ –õ–û–ö–ê–õ–¨–ù–´–ô –†–ï–ñ–ò–ú: Google Colab –æ–±–Ω–∞—Ä—É–∂–µ–Ω")
except ImportError:
    print("‚ùå –û–®–ò–ë–ö–ê: –°–∏—Å—Ç–µ–º–∞ –ø—Ä–µ–¥–Ω–∞–∑–Ω–∞—á–µ–Ω–∞ —Ç–æ–ª—å–∫–æ –¥–ª—è Google Colab!")
    raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –ó–∞–ø—É—Å–∫ –≤–æ–∑–º–æ–∂–µ–Ω —Ç–æ–ª—å–∫–æ –≤ Google Colab!")

print("üöÄ –ù–ê–ß–ò–ù–ê–Æ –£–°–¢–ê–ù–û–í–ö–£ –õ–û–ö–ê–õ–¨–ù–´–• –ó–ê–í–ò–°–ò–ú–û–°–¢–ï–ô...")
print("üîí –ü–†–ò–í–ê–¢–ù–û–°–¢–¨: –í—Å–µ –º–æ–¥–µ–ª–∏ —Ä–∞–±–æ—Ç–∞—é—Ç –ª–æ–∫–∞–ª—å–Ω–æ!")

# –°–ø–∏—Å–æ–∫ –ö–†–ò–¢–ò–ß–ï–°–ö–ò–• –õ–û–ö–ê–õ–¨–ù–´–• –ø–∞–∫–µ—Ç–æ–≤
CRITICAL_PACKAGES = [
    "torch>=2.0.0 torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118",
    "openai-whisper>=20231117",
    "transformers>=4.30.0",
    "librosa>=0.10.0",
    "soundfile>=0.12.0",
    "pydub>=0.25.0",
    "ffmpeg-python",
    "scipy",
    "noisereduce",
    "nltk>=3.8.0",
    "pymorphy2>=0.9.0", 
    "spacy>=3.6.0",
    "sentence-transformers>=2.2.0",
    "jiwer>=3.0.0",
    "scikit-learn>=1.3.0",
    "evaluate",
    "python-dotenv",
    "tqdm",
    "colorama"
]

def critical_install(package):
    """–ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–∞–∫–µ—Ç–∞ - –ø—Ä–∏ –æ—à–∏–±–∫–µ —Å–∏—Å—Ç–µ–º–∞ –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ—Ç—Å—è"""
    print(f"üîß –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—é: {package.split()[0]}")
    
    try:
        result = subprocess.run(
            f"pip install --quiet --upgrade {package}", 
            shell=True, 
            capture_output=True, 
            text=True,
            timeout=300
        )
        
        if result.returncode != 0:
            print(f"‚ùå –û–®–ò–ë–ö–ê —É—Å—Ç–∞–Ω–æ–≤–∫–∏: {package}")
            print(f"STDERR: {result.stderr}")
            raise RuntimeError(f"‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å {package}")
        
        print(f"‚úÖ –£–°–ü–ï–®–ù–û: {package.split()[0]}")
        
    except subprocess.TimeoutExpired:
        raise RuntimeError(f"‚ùå –ü—Ä–µ–≤—ã—à–µ–Ω–æ –≤—Ä–µ–º—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏ {package}")
    except Exception as e:
        raise RuntimeError(f"‚ùå –û–®–ò–ë–ö–ê —É—Å—Ç–∞–Ω–æ–≤–∫–∏ {package}: {e}")

# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–∞–∫–µ—Ç–æ–≤
start_time = time.time()
for package in CRITICAL_PACKAGES:
    critical_install(package)

# –°–∏—Å—Ç–µ–º–Ω—ã–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏
print("üîß –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—é —Å–∏—Å—Ç–µ–º–Ω—ã–µ –ø–∞–∫–µ—Ç—ã...")
system_result = subprocess.run("apt-get update -qq && apt-get install -qq -y ffmpeg sox libsox-fmt-all", shell=True, capture_output=True, text=True)

if system_result.returncode != 0:
    raise RuntimeError(f"‚ùå –û—à–∏–±–∫–∞ —Å–∏—Å—Ç–µ–º–Ω—ã—Ö –ø–∞–∫–µ—Ç–æ–≤: {system_result.stderr}")

install_time = time.time() - start_time
print(f"\n‚úÖ –í–°–ï –õ–û–ö–ê–õ–¨–ù–´–ï –ó–ê–í–ò–°–ò–ú–û–°–¢–ò –£–°–¢–ê–ù–û–í–õ–ï–ù–´!")
print(f"‚è±Ô∏è –í—Ä–µ–º—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏: {install_time:.1f}—Å")
print(f"üîí –ü–†–ò–í–ê–¢–ù–û–°–¢–¨: –ì–æ—Ç–æ–≤ –∫ –ª–æ–∫–∞–ª—å–Ω–æ–π —Ä–∞–±–æ—Ç–µ!")
```

```python
# üö® –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ô –ò–ú–ü–û–†–¢ –õ–û–ö–ê–õ–¨–ù–´–• –ö–û–ú–ü–û–ù–ï–ù–¢–û–í
import sys
import warnings
warnings.filterwarnings("ignore")

print("üö® –õ–û–ö–ê–õ–¨–ù–´–ô –ò–ú–ü–û–†–¢ - –ù–ò–ö–ê–ö–ò–• API!")
print("üîí –ü–†–ò–í–ê–¢–ù–û–°–¢–¨: –¢–æ–ª—å–∫–æ –ª–æ–∫–∞–ª—å–Ω—ã–µ –º–æ–¥–µ–ª–∏!")
print("="*60)

# –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ô —Å–ø–∏—Å–æ–∫ –õ–û–ö–ê–õ–¨–ù–´–• –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤
CRITICAL_IMPORTS = [
    ('enhanced_transcriber', 'EnhancedTranscriber', '–û—Å–Ω–æ–≤–Ω–æ–π –∫–ª–∞—Å—Å —Å–∏—Å—Ç–µ–º—ã'),
    ('core.interfaces.transcriber', 'ITranscriber', '–ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å —Ç—Ä–∞–Ω—Å–∫—Ä–∞–π–±–µ—Ä–∞'),
    ('core.interfaces.audio_processor', 'IAudioProcessor', '–ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å –∞—É–¥–∏–æ –ø—Ä–æ—Ü–µ—Å—Å–æ—Ä–∞'),
    ('core.interfaces.quality_assessor', 'IQualityAssessor', '–ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞'),
    ('providers.whisper.whisper_local', 'WhisperLocalTranscriber', 'Whisper Local —Ç—Ä–∞–Ω—Å–∫—Ä–∞–π–±–µ—Ä'),
    ('services.audio_processor', 'AudioProcessorService', '–°–µ—Ä–≤–∏—Å –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∞—É–¥–∏–æ'),
    ('services.ensemble_service', 'EnsembleTranscriptionService', '–°–µ—Ä–≤–∏—Å –∞–Ω—Å–∞–º–±–ª—è'),
    ('services.quality_assessor', 'QualityAssessmentService', '–°–µ—Ä–≤–∏—Å –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞')
]

# –ò–ú–ü–û–†–¢ –∫–∞–∂–¥–æ–≥–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞
imported_components = {}
import_errors = []

for module_name, class_name, description in CRITICAL_IMPORTS:
    print(f"üîÑ –ò–º–ø–æ—Ä—Ç: {module_name}.{class_name}")
    print(f"   üìù {description}")
    
    try:
        module = __import__(module_name, fromlist=[class_name])
        component_class = getattr(module, class_name)
        imported_components[class_name] = component_class
        print(f"   ‚úÖ –£–°–ü–ï–®–ù–û: {class_name}")
        
    except Exception as e:
        error_msg = f"‚ùå –û–®–ò–ë–ö–ê {module_name}.{class_name}: {e}"
        print(f"   {error_msg}")
        import_errors.append(error_msg)
    
    print()

# T-one –ø—Ä–æ–≤–∞–π–¥–µ—Ä (–æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–π)
print("üîÑ –ü—Ä–æ–≤–µ—Ä–∫–∞ T-one –ø—Ä–æ–≤–∞–π–¥–µ—Ä–∞...")
try:
    from providers.tone.tone_provider import ToneTranscriber
    imported_components['ToneTranscriber'] = ToneTranscriber
    print("   ‚úÖ T-one –¥–æ—Å—Ç—É–ø–µ–Ω!")
    TONE_AVAILABLE = True
except ImportError as e:
    print(f"   ‚ö†Ô∏è T-one –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω: {e}")
    TONE_AVAILABLE = False

print("="*60)

if import_errors:
    print("‚ùå –û–®–ò–ë–ö–ò –ò–ú–ü–û–†–¢–ê:")
    for error in import_errors:
        print(f"   {error}")
    raise RuntimeError("‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã!")

print("üéâ –í–°–ï –õ–û–ö–ê–õ–¨–ù–´–ï –ö–û–ú–ü–û–ù–ï–ù–¢–´ –ó–ê–ì–†–£–ñ–ï–ù–´!")
print(f"‚úÖ –ö–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤: {len(imported_components)}")
print(f"ü§ñ T-one: {'–î–æ—Å—Ç—É–ø–µ–Ω' if TONE_AVAILABLE else '–ù–µ–¥–æ—Å—Ç—É–ø–µ–Ω'}")
print("üîí –ü–†–ò–í–ê–¢–ù–û–°–¢–¨: –í—Å–µ –º–æ–¥–µ–ª–∏ –ª–æ–∫–∞–ª—å–Ω—ã–µ!")
```
