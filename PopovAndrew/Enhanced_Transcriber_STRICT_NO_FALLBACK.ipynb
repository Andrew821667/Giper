````python
# üö® –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –£–°–¢–ê–ù–û–í–ö–ê –í–°–ï–• –ó–ê–í–ò–°–ò–ú–û–°–¢–ï–ô
import os
import sys
import subprocess
from pathlib import Path
import time

# –°–¢–†–û–ì–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ä–µ–¥—ã
try:
    import google.colab
    IN_COLAB = True
    print("üéØ –°–¢–†–û–ì–ò–ô –†–ï–ñ–ò–ú: Google Colab –æ–±–Ω–∞—Ä—É–∂–µ–Ω")
except ImportError:
    print("‚ùå –û–®–ò–ë–ö–ê: –°–∏—Å—Ç–µ–º–∞ –ø—Ä–µ–¥–Ω–∞–∑–Ω–∞—á–µ–Ω–∞ —Ç–æ–ª—å–∫–æ –¥–ª—è Google Colab!")
    raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –ó–∞–ø—É—Å–∫ –≤–æ–∑–º–æ–∂–µ–Ω —Ç–æ–ª—å–∫–æ –≤ Google Colab!")

print("üöÄ –ù–ê–ß–ò–ù–ê–Æ –ö–†–ò–¢–ò–ß–ï–°–ö–£–Æ –£–°–¢–ê–ù–û–í–ö–£ –ó–ê–í–ò–°–ò–ú–û–°–¢–ï–ô...")
print("‚ö†Ô∏è  –í–ù–ò–ú–ê–ù–ò–ï: –õ—é–±–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–∏–≤–µ–¥–µ—Ç –∫ –æ—Å—Ç–∞–Ω–æ–≤–∫–µ —Å–∏—Å—Ç–µ–º—ã!")

# –°–ø–∏—Å–æ–∫ –ö–†–ò–¢–ò–ß–ï–°–ö–ò–• –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–µ–π - –í–°–ï –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω—ã
CRITICAL_PACKAGES = [
    "torch>=2.0.0 torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118",
    "openai-whisper>=20231117",
    "openai>=1.0.0",
    "transformers>=4.30.0",
    "librosa>=0.10.0",
    "soundfile>=0.12.0",
    "pydub>=0.25.0",
    "ffmpeg-python",
    "moviepy",
    "scipy",
    "noisereduce",
    "nltk>=3.8.0",
    "pymorphy2>=0.9.0", 
    "spacy>=3.6.0",
    "sentence-transformers>=2.2.0",
    "jiwer>=3.0.0",
    "scikit-learn>=1.3.0",
    "evaluate",
    "python-dotenv",
    "tqdm",
    "colorama",
    "yt-dlp"
]

def critical_install(package):
    """–ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø–∞–∫–µ—Ç–∞ - –ø—Ä–∏ –æ—à–∏–±–∫–µ —Å–∏—Å—Ç–µ–º–∞ –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ—Ç—Å—è"""
    print(f"üîß –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—é –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π –ø–∞–∫–µ—Ç: {package.split()[0]}")
    
    try:
        result = subprocess.run(
            f"pip install --quiet --upgrade {package}", 
            shell=True, 
            capture_output=True, 
            text=True,
            timeout=300  # 5 –º–∏–Ω—É—Ç –º–∞–∫—Å–∏–º—É–º –Ω–∞ –ø–∞–∫–µ—Ç
        )
        
        if result.returncode != 0:
            print(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê —É—Å—Ç–∞–Ω–æ–≤–∫–∏: {package}")
            print(f"STDERR: {result.stderr}")
            raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –ù–µ —É–¥–∞–ª–æ—Å—å —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å {package}")
        
        print(f"‚úÖ –£–°–ü–ï–®–ù–û: {package.split()[0]}")
        
    except subprocess.TimeoutExpired:
        raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –ü—Ä–µ–≤—ã—à–µ–Ω–æ –≤—Ä–µ–º—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏ {package}")
    except Exception as e:
        raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê —É—Å—Ç–∞–Ω–æ–≤–∫–∏ {package}: {e}")

# –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∫–∞–∂–¥–æ–≥–æ –ø–∞–∫–µ—Ç–∞
start_time = time.time()

for package in CRITICAL_PACKAGES:
    critical_install(package)

# –°–∏—Å—Ç–µ–º–Ω—ã–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ - —Ç–æ–∂–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ
print("üîß –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—é –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ —Å–∏—Å—Ç–µ–º–Ω—ã–µ –ø–∞–∫–µ—Ç—ã...")
system_result = subprocess.run("apt-get update -qq && apt-get install -qq -y ffmpeg sox libsox-fmt-all", shell=True, capture_output=True, text=True)

if system_result.returncode != 0:
    raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê —Å–∏—Å—Ç–µ–º–Ω—ã—Ö –ø–∞–∫–µ—Ç–æ–≤: {system_result.stderr}")

install_time = time.time() - start_time
print(f"\n‚úÖ –í–°–ï –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï –ó–ê–í–ò–°–ò–ú–û–°–¢–ò –£–°–¢–ê–ù–û–í–õ–ï–ù–´!")
print(f"‚è±Ô∏è  –í—Ä–µ–º—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏: {install_time:.1f}—Å")
print(f"üì¶ –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–æ –ø–∞–∫–µ—Ç–æ–≤: {len(CRITICAL_PACKAGES) + 1}")
print("üéØ –°–ò–°–¢–ï–ú–ê –ì–û–¢–û–í–ê –ö –°–¢–†–û–ì–û–ô –ü–†–û–í–ï–†–ö–ï –ö–û–ú–ü–û–ù–ï–ù–¢–û–í!")
````

````python
# üö® –ö–†–ò–¢–ò–ß–ï–°–ö–û–ï —Å–∫–∞—á–∏–≤–∞–Ω–∏–µ —è–∑—ã–∫–æ–≤—ã—Ö –º–æ–¥–µ–ª–µ–π
import nltk
import subprocess

print("üöÄ –ö–†–ò–¢–ò–ß–ï–°–ö–û–ï —Å–∫–∞—á–∏–≤–∞–Ω–∏–µ —è–∑—ã–∫–æ–≤—ã—Ö —Ä–µ—Å—É—Ä—Å–æ–≤...")

# –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï NLTK —Ä–µ—Å—É—Ä—Å—ã
CRITICAL_NLTK = [
    'punkt', 'stopwords', 'averaged_perceptron_tagger',
    'wordnet', 'omw-1.4', 'vader_lexicon'
]

def critical_nltk_download(resource):
    """–ö–†–ò–¢–ò–ß–ï–°–ö–û–ï —Å–∫–∞—á–∏–≤–∞–Ω–∏–µ NLTK —Ä–µ—Å—É—Ä—Å–∞"""
    try:
        nltk.download(resource, quiet=True)
        print(f"‚úÖ NLTK {resource} - –∑–∞–≥—Ä—É–∂–µ–Ω")
    except Exception as e:
        raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –∑–∞–≥—Ä—É–∑–∫–∏ NLTK {resource}: {e}")

for resource in CRITICAL_NLTK:
    critical_nltk_download(resource)

# –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø SpaCy —Ä—É—Å—Å–∫–∞—è –º–æ–¥–µ–ª—å
print("üá∑üá∫ –ó–∞–≥—Ä—É–∂–∞—é –ö–†–ò–¢–ò–ß–ï–°–ö–£–Æ —Ä—É—Å—Å–∫—É—é –º–æ–¥–µ–ª—å SpaCy...")

try:
    spacy_result = subprocess.run(
        "python -m spacy download ru_core_news_sm", 
        shell=True, 
        capture_output=True, 
        text=True,
        timeout=300
    )
    
    if spacy_result.returncode != 0:
        raise RuntimeError(f"‚ùå –û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ SpaCy –º–æ–¥–µ–ª–∏: {spacy_result.stderr}")
    
    print("‚úÖ SpaCy —Ä—É—Å—Å–∫–∞—è –º–æ–¥–µ–ª—å –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –£–°–ü–ï–®–ù–û")
    
except subprocess.TimeoutExpired:
    raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –ü—Ä–µ–≤—ã—à–µ–Ω–æ –≤—Ä–µ–º—è –∑–∞–≥—Ä—É–∑–∫–∏ SpaCy –º–æ–¥–µ–ª–∏")
except Exception as e:
    raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –∑–∞–≥—Ä—É–∑–∫–∏ SpaCy: {e}")

print("\n‚úÖ –í–°–ï –Ø–ó–´–ö–û–í–´–ï –ú–û–î–ï–õ–ò –ó–ê–ì–†–£–ñ–ï–ù–´ –ö–†–ò–¢–ò–ß–ï–°–ö–ò –£–°–ü–ï–®–ù–û!")
print("üéØ –°–ò–°–¢–ï–ú–ê –ì–û–¢–û–í–ê –ö –ö–õ–û–ù–ò–†–û–í–ê–ù–ò–Æ –ü–†–û–ï–ö–¢–ê!")
````

````python
# üö® –ö–†–ò–¢–ò–ß–ï–°–ö–û–ï –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ –ü–û–õ–ù–û–ì–û –ø—Ä–æ–µ–∫—Ç–∞
import shutil
import subprocess
from pathlib import Path

print("üöÄ –ö–†–ò–¢–ò–ß–ï–°–ö–û–ï –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ Enhanced Transcriber –ø—Ä–æ–µ–∫—Ç–∞...")

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ö–†–ò–¢–ò–ß–ï–°–ö–ò–• –ø—É—Ç–µ–π
WORK_DIR = Path('/content')
REPO_DIR = WORK_DIR / 'Giper'
PROJECT_DIR = REPO_DIR / 'PopovAndrew' / 'enhanced-transcriber'

print(f"üìÅ –¶–µ–ª–µ–≤–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è: {PROJECT_DIR}")

# –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –æ—á–∏—Å—Ç–∫–∞
if REPO_DIR.exists():
    print("üóëÔ∏è  –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –æ—á–∏—Å—Ç–∫–∞ —Å—Ç–∞—Ä–æ–≥–æ —Ä–µ–ø–æ–∑–∏—Ç–æ—Ä–∏—è...")
    try:
        shutil.rmtree(REPO_DIR)
        print("‚úÖ –°—Ç–∞—Ä—ã–π —Ä–µ–ø–æ–∑–∏—Ç–æ—Ä–∏–π —É–¥–∞–ª–µ–Ω")
    except Exception as e:
        raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –æ—á–∏—Å—Ç–∫–∏: {e}")

# –ö–†–ò–¢–ò–ß–ï–°–ö–û–ï –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ
os.chdir(WORK_DIR)
print("üì° –í—ã–ø–æ–ª–Ω—è—é –ö–†–ò–¢–ò–ß–ï–°–ö–û–ï –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ —Ä–µ–ø–æ–∑–∏—Ç–æ—Ä–∏—è...")

try:
    clone_result = subprocess.run(
        "git clone https://github.com/Andrew821667/Giper.git",
        shell=True,
        capture_output=True,
        text=True,
        timeout=120  # 2 –º–∏–Ω—É—Ç—ã –º–∞–∫—Å–∏–º—É–º
    )
    
    if clone_result.returncode != 0:
        raise RuntimeError(f"‚ùå –û—à–∏–±–∫–∞ –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è: {clone_result.stderr}")
    
    print("‚úÖ –†–µ–ø–æ–∑–∏—Ç–æ—Ä–∏–π –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω –£–°–ü–ï–®–ù–û")
    
except subprocess.TimeoutExpired:
    raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –ü—Ä–µ–≤—ã—à–µ–Ω–æ –≤—Ä–µ–º—è –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è")
except Exception as e:
    raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –∫–ª–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è: {e}")

# –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –ø—Ä–æ–µ–∫—Ç–∞
REQUIRED_FILES = [
    'enhanced_transcriber.py',
    'main.py',
    'requirements.txt',
    'core/interfaces/transcriber.py',
    'providers/whisper/whisper_local.py',
    'providers/whisper/whisper_openai.py',
    'services/audio_processor.py',
    'services/ensemble_service.py',
    'services/quality_assessor.py'
]

REQUIRED_DIRS = [
    'core/interfaces',
    'providers/whisper',
    'providers/tone', 
    'services'
]

print("\nüîç –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –ø—Ä–æ–µ–∫—Ç–∞...")

if not PROJECT_DIR.exists():
    raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –ø—Ä–æ–µ–∫—Ç–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞: {PROJECT_DIR}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ö–†–ò–¢–ò–ß–ï–°–ö–ò–• –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–π
missing_dirs = []
for dir_path in REQUIRED_DIRS:
    full_path = PROJECT_DIR / dir_path
    if not full_path.exists():
        missing_dirs.append(dir_path)
    else:
        print(f"‚úÖ –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –Ω–∞–π–¥–µ–Ω–∞: {dir_path}")

if missing_dirs:
    raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –û—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏: {missing_dirs}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ö–†–ò–¢–ò–ß–ï–°–ö–ò–• —Ñ–∞–π–ª–æ–≤
missing_files = []
for file_path in REQUIRED_FILES:
    full_path = PROJECT_DIR / file_path
    if not full_path.exists():
        missing_files.append(file_path)
    else:
        print(f"‚úÖ –§–∞–π–ª –Ω–∞–π–¥–µ–Ω: {file_path}")

if missing_files:
    raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –û—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï —Ñ–∞–π–ª—ã: {missing_files}")

# –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –ø—É—Ç–µ–π
os.chdir(PROJECT_DIR)
if str(PROJECT_DIR) not in sys.path:
    sys.path.insert(0, str(PROJECT_DIR))

print(f"\n‚úÖ –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –ü–†–û–í–ï–†–ö–ê –ó–ê–í–ï–†–®–ï–ù–ê –£–°–ü–ï–®–ù–û!")
print(f"üìÅ –†–∞–±–æ—á–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è: {PROJECT_DIR}")
print(f"üìä –ü—Ä–æ–≤–µ—Ä–µ–Ω–æ —Ñ–∞–π–ª–æ–≤: {len(REQUIRED_FILES)}")
print(f"üìä –ü—Ä–æ–≤–µ—Ä–µ–Ω–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–π: {len(REQUIRED_DIRS)}")
print("üéØ –°–ò–°–¢–ï–ú–ê –ì–û–¢–û–í–ê –ö –°–¢–†–û–ì–û–ú–£ –ò–ú–ü–û–†–¢–£!")
````

````python
# ‚ö° –°–¢–†–û–ì–ê–Ø –ò–ù–ò–¶–ò–ê–õ–ò–ó–ê–¶–ò–Ø –ü–û–õ–ù–û–ô –°–ò–°–¢–ï–ú–´
import torch
from typing import List, Dict, Optional
from datetime import datetime

class StrictEnhancedTranscriber:
    """–°–¢–†–û–ì–ê–Ø –≤–µ—Ä—Å–∏—è Enhanced Transcriber - –ë–ï–ó –ö–û–ú–ü–†–û–ú–ò–°–°–û–í!"""
    
    def __init__(self):
        print("‚ö° –°–¢–†–û–ì–ê–Ø –ò–ù–ò–¶–ò–ê–õ–ò–ó–ê–¶–ò–Ø ENHANCED TRANSCRIBER")
        print("üö® –†–ï–ñ–ò–ú: –ù–ò–ö–ê–ö–ò–• FALLBACK - –¢–û–õ–¨–ö–û –ü–û–õ–ù–ê–Ø –°–ò–°–¢–ï–ú–ê!")
        print("="*60)
        
        # –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–∞
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        if not torch.cuda.is_available():
            print("‚ö†Ô∏è  –ü–†–ï–î–£–ü–†–ï–ñ–î–ï–ù–ò–ï: GPU –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω - –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –±—É–¥–µ—Ç —Å–Ω–∏–∂–µ–Ω–∞!")
        
        print(f"üíª –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {self.device}")
        
        if torch.cuda.is_available():
            gpu_props = torch.cuda.get_device_properties(0)
            print(f"üéÆ GPU: {gpu_props.name}")
            print(f"üíæ GPU –ø–∞–º—è—Ç—å: {gpu_props.total_memory // 1024**3} GB")
        
        # –°–¢–†–û–ì–ê–Ø –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–æ–≤–∞–π–¥–µ—Ä–æ–≤
        self.providers = {}
        self._init_providers_strict()
        
        # –°–¢–†–û–ì–ê–Ø –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å–µ—Ä–≤–∏—Å–æ–≤
        self._init_services_strict()
        
        # –§–ò–ù–ê–õ–¨–ù–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å–∏—Å—Ç–µ–º—ã
        self._final_system_check()
        
        print("\nüéâ –°–¢–†–û–ì–ê–Ø –°–ò–°–¢–ï–ú–ê –ò–ù–ò–¶–ò–ê–õ–ò–ó–ò–†–û–í–ê–ù–ê –£–°–ü–ï–®–ù–û!")
        print("üéØ –ì–ê–†–ê–ù–¢–ò–Ø –ö–ê–ß–ï–°–¢–í–ê: 95%+ –î–õ–Ø –†–£–°–°–ö–û–ì–û –Ø–ó–´–ö–ê")
        print("‚ö° –ü–û–õ–ù–ê–Ø –ú–û–©–ù–û–°–¢–¨ ENHANCED TRANSCRIBER –ê–ö–¢–ò–í–ò–†–û–í–ê–ù–ê!")
    
    def _init_providers_strict(self):
        """–°–¢–†–û–ì–ê–Ø –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –≤—Å–µ—Ö –ø—Ä–æ–≤–∞–π–¥–µ—Ä–æ–≤"""
        print("\nü§ñ –°–¢–†–û–ì–ê–Ø –ò–ù–ò–¶–ò–ê–õ–ò–ó–ê–¶–ò–Ø –ü–†–û–í–ê–ô–î–ï–†–û–í...")
        
        provider_errors = []
        
        # 1. Whisper Local - –û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–´–ô
        print("üîÑ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Whisper Local (–ö–†–ò–¢–ò–ß–ï–°–ö–ò–ô)...")
        try:
            WhisperLocalProvider = imported_components['WhisperLocalProvider']
            self.providers['whisper_local'] = WhisperLocalProvider(
                model_size='base',
                device=self.device
            )
            print("‚úÖ Whisper Local - –ì–û–¢–û–í")
        except Exception as e:
            error_msg = f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê Whisper Local: {e}"
            print(error_msg)
            provider_errors.append(error_msg)
        
        # 2. Whisper OpenAI - –£–°–õ–û–í–ù–û –û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–´–ô
        if OPENAI_AVAILABLE:
            print("üîÑ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Whisper OpenAI...")
            try:
                WhisperOpenAIProvider = imported_components['WhisperOpenAIProvider']
                self.providers['whisper_openai'] = WhisperOpenAIProvider(
                    api_key=os.environ['OPENAI_API_KEY']
                )
                print("‚úÖ Whisper OpenAI - –ì–û–¢–û–í")
            except Exception as e:
                print(f"‚ö†Ô∏è  Whisper OpenAI –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω: {e}")
        else:
            print("‚ö†Ô∏è  Whisper OpenAI –ø—Ä–æ–ø—É—â–µ–Ω - –Ω–µ—Ç API –∫–ª—é—á–∞")
        
        # 3. T-one - –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ô
        if TONE_AVAILABLE:
            print("üîÑ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è T-one –ø—Ä–æ–≤–∞–π–¥–µ—Ä–∞...")
            try:
                ToneProvider = imported_components['ToneProvider']
                self.providers['tone'] = ToneProvider()
                print("‚úÖ T-one - –ì–û–¢–û–í")
            except Exception as e:
                print(f"‚ö†Ô∏è  T-one –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω: {e}")
        else:
            print("‚ö†Ô∏è  T-one –ø—Ä–æ–≤–∞–π–¥–µ—Ä –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω")
        
        # –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–≤–∞–π–¥–µ—Ä–æ–≤
        if not self.providers:
            raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –ù–ò –û–î–ò–ù –ü–†–û–í–ê–ô–î–ï–† –ù–ï –ò–ù–ò–¶–ò–ê–õ–ò–ó–ò–†–û–í–ê–ù!")
        
        if 'whisper_local' not in self.providers:
            raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: Whisper Local –û–ë–Ø–ó–ê–¢–ï–õ–ï–ù –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å–∏—Å—Ç–µ–º—ã!")
        
        print(f"üéØ –ê–ö–¢–ò–í–ù–´–• –ü–†–û–í–ê–ô–î–ï–†–û–í: {len(self.providers)}")
        for name in self.providers.keys():
            print(f"   ‚úÖ {name}")
    
    def _init_services_strict(self):
        """–°–¢–†–û–ì–ê–Ø –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –≤—Å–µ—Ö —Å–µ—Ä–≤–∏—Å–æ–≤"""
        print("\nüîß –°–¢–†–û–ì–ê–Ø –ò–ù–ò–¶–ò–ê–õ–ò–ó–ê–¶–ò–Ø –°–ï–†–í–ò–°–û–í...")
        
        service_errors = []
        
        # 1. Audio Processor - –û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–´–ô
        print("üîÑ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Audio Processor (–ö–†–ò–¢–ò–ß–ï–°–ö–ò–ô)...")
        try:
            AudioProcessor = imported_components['AudioProcessor']
            self.audio_processor = AudioProcessor()
            print("‚úÖ Audio Processor - –ì–û–¢–û–í")
        except Exception as e:
            error_msg = f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê Audio Processor: {e}"
            print(error_msg)
            service_errors.append(error_msg)
        
        # 2. Quality Assessor - –û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–´–ô
        print("üîÑ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Quality Assessor (–ö–†–ò–¢–ò–ß–ï–°–ö–ò–ô)...")
        try:
            QualityAssessor = imported_components['QualityAssessor']
            self.quality_assessor = QualityAssessor()
            print("‚úÖ Quality Assessor - –ì–û–¢–û–í")
        except Exception as e:
            error_msg = f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê Quality Assessor: {e}"
            print(error_msg)
            service_errors.append(error_msg)
        
        # 3. Ensemble Service - –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ô –¥–ª—è –ø–æ–ª–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã
        if len(self.providers) > 1:
            print("üîÑ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Ensemble Service (–ö–†–ò–¢–ò–ß–ï–°–ö–ò–ô)...")
            try:
                EnsembleService = imported_components['EnsembleService']
                self.ensemble_service = EnsembleService(
                    providers=list(self.providers.values())
                )
                print("‚úÖ Ensemble Service - –ì–û–¢–û–í")
            except Exception as e:
                error_msg = f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê Ensemble Service: {e}"
                print(error_msg)
                service_errors.append(error_msg)
        else:
            self.ensemble_service = None
            print("‚ö†Ô∏è  Ensemble Service –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω - –Ω—É–∂–Ω–æ >1 –ø—Ä–æ–≤–∞–π–¥–µ—Ä–∞")
        
        # –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Å–µ—Ä–≤–∏—Å–æ–≤
        if service_errors:
            print("\n‚ùå‚ùå‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï –û–®–ò–ë–ö–ò –°–ï–†–í–ò–°–û–í! ‚ùå‚ùå‚ùå")
            for error in service_errors:
                print(f"   {error}")
            raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï –°–ï–†–í–ò–°–´ –ù–ï –ò–ù–ò–¶–ò–ê–õ–ò–ó–ò–†–û–í–ê–ù–´!")
        
        print("üéØ –í–°–ï –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï –°–ï–†–í–ò–°–´ –ò–ù–ò–¶–ò–ê–õ–ò–ó–ò–†–û–í–ê–ù–´!")
    
    def _final_system_check(self):
        """–§–ò–ù–ê–õ–¨–ù–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Ü–µ–ª–æ—Å—Ç–Ω–æ—Å—Ç–∏ —Å–∏—Å—Ç–µ–º—ã"""
        print("\nüîç –§–ò–ù–ê–õ–¨–ù–ê–Ø –ü–†–û–í–ï–†–ö–ê –¶–ï–õ–û–°–¢–ù–û–°–¢–ò –°–ò–°–¢–ï–ú–´...")
        
        checks = [
            (len(self.providers) >= 1, "–ú–∏–Ω–∏–º—É–º 1 –ø—Ä–æ–≤–∞–π–¥–µ—Ä"),
            ('whisper_local' in self.providers, "Whisper Local –ø—Ä–æ–≤–∞–π–¥–µ—Ä"),
            (hasattr(self, 'audio_processor'), "Audio Processor —Å–µ—Ä–≤–∏—Å"),
            (hasattr(self, 'quality_assessor'), "Quality Assessor —Å–µ—Ä–≤–∏—Å"),
            (self.audio_processor is not None, "Audio Processor –∞–∫—Ç–∏–≤–µ–Ω"),
            (self.quality_assessor is not None, "Quality Assessor –∞–∫—Ç–∏–≤–µ–Ω")
        ]
        
        failed_checks = []
        
        for check_result, check_name in checks:
            if check_result:
                print(f"‚úÖ {check_name}")
            else:
                print(f"‚ùå {check_name}")
                failed_checks.append(check_name)
        
        if failed_checks:
            print("\nüí• –°–ò–°–¢–ï–ú–ê –ù–ï –ü–†–û–®–õ–ê –§–ò–ù–ê–õ–¨–ù–£–Æ –ü–†–û–í–ï–†–ö–£!")
            print("‚ùå –ù–µ—É–¥–∞–≤—à–∏–µ—Å—è –ø—Ä–æ–≤–µ—Ä–∫–∏:")
            for check in failed_checks:
                print(f"   ‚Ä¢ {check}")
            raise RuntimeError("‚ùå –°–ò–°–¢–ï–ú–ê –ù–ï –ì–û–¢–û–í–ê –ö –†–ê–ë–û–¢–ï!")
        
        print("\nüéâ –°–ò–°–¢–ï–ú–ê –ü–†–û–®–õ–ê –í–°–ï –ü–†–û–í–ï–†–ö–ò!")
    
    def transcribe_professional_strict(self, 
                                     audio_path: str,
                                     language: str = "ru",
                                     force_ensemble: bool = True,
                                     force_preprocessing: bool = True,
                                     force_quality_check: bool = True) -> Dict:
        """
        –°–¢–†–û–ì–ê–Ø –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω–∞—è —Ç—Ä–∞–Ω—Å–∫—Ä–∏–ø—Ü–∏—è
        –ë–ï–ó –ö–û–ú–ü–†–û–ú–ò–°–°–û–í - –í–°–ï –≠–¢–ê–ü–´ –û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–´!
        """
        
        print(f"‚ö° –°–¢–†–û–ì–ê–Ø –ü–†–û–§–ï–°–°–ò–û–ù–ê–õ–¨–ù–ê–Ø –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–Ø")
        print(f"üìÅ –§–∞–π–ª: {Path(audio_path).name}")
        print(f"üåç –Ø–∑—ã–∫: {language}")
        print(f"ü§ñ –ê–Ω—Å–∞–º–±–ª—å: {'–ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–û' if force_ensemble else '–û–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ'}")
        print(f"üéµ –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞: {'–ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–û' if force_preprocessing else '–û–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ'}")
        print(f"üìä –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞: {'–ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–û' if force_quality_check else '–û–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ'}")
        print("="*50)
        
        start_time = datetime.now()
        
        # –≠–¢–ê–ü 1: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–ê–Ø –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞
        if force_preprocessing:
            print("üéµ –≠–¢–ê–ü 1: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–ê–Ø –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –∞—É–¥–∏–æ...")
            try:
                processed_audio = self.audio_processor.process_file(
                    audio_path,
                    enhance_speech=True,
                    reduce_noise=True,
                    normalize=True
                )
                print("‚úÖ –≠–¢–ê–ü 1 –ó–ê–í–ï–†–®–ï–ù: –ê—É–¥–∏–æ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∞–Ω–æ")
            except Exception as e:
                raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –≠–¢–ê–ü–ê 1: {e}")
        else:
            processed_audio = audio_path
            print("‚ö†Ô∏è  –≠–¢–ê–ü 1 –ü–†–û–ü–£–©–ï–ù: –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –æ—Ç–∫–ª—é—á–µ–Ω–∞")
        
        # –≠–¢–ê–ü 2: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–´–ô –∞–Ω—Å–∞–º–±–ª—å –∏–ª–∏ –æ–¥–∏–Ω–æ—á–Ω–∞—è –º–æ–¥–µ–ª—å
        if force_ensemble and self.ensemble_service and len(self.providers) > 1:
            print("ü§ñ –≠–¢–ê–ü 2: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–´–ô –∞–Ω—Å–∞–º–±–ª—å –º–æ–¥–µ–ª–µ–π...")
            try:
                result = self.ensemble_service.transcribe(
                    processed_audio,
                    language=language
                )
                print(f"‚úÖ –≠–¢–ê–ü 2 –ó–ê–í–ï–†–®–ï–ù: –ê–Ω—Å–∞–º–±–ª—å ({len(self.providers)} –º–æ–¥–µ–ª–µ–π)")
                transcription_method = f"ensemble_{len(self.providers)}_models"
            except Exception as e:
                raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –≠–¢–ê–ü–ê 2 (–∞–Ω—Å–∞–º–±–ª—å): {e}")
        else:
            print("ü§ñ –≠–¢–ê–ü 2: –û–¥–∏–Ω–æ—á–Ω–∞—è –º–æ–¥–µ–ª—å (–∞–Ω—Å–∞–º–±–ª—å –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω)...")
            try:
                primary_provider = self.providers['whisper_local']
                result = primary_provider.transcribe(
                    processed_audio,
                    language=language
                )
                print("‚úÖ –≠–¢–ê–ü 2 –ó–ê–í–ï–†–®–ï–ù: –û–¥–∏–Ω–æ—á–Ω–∞—è –º–æ–¥–µ–ª—å")
                transcription_method = "single_whisper_local"
            except Exception as e:
                raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –≠–¢–ê–ü–ê 2 (–æ–¥–∏–Ω–æ—á–Ω–∞—è): {e}")
        
        # –≠–¢–ê–ü 3: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–ê–Ø –æ—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞
        if force_quality_check:
            print("üìä –≠–¢–ê–ü 3: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–ê–Ø –æ—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞...")
            try:
                quality_metrics = self.quality_assessor.assess(
                    result['text'],
                    audio_path=processed_audio,
                    language=language
                )
                print("‚úÖ –≠–¢–ê–ü 3 –ó–ê–í–ï–†–®–ï–ù: –ö–∞—á–µ—Å—Ç–≤–æ –æ—Ü–µ–Ω–µ–Ω–æ")
            except Exception as e:
                raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –≠–¢–ê–ü–ê 3: {e}")
        else:
            quality_metrics = {'confidence_score': 0.0, 'note': 'quality_check_disabled'}
            print("‚ö†Ô∏è  –≠–¢–ê–ü 3 –ü–†–û–ü–£–©–ï–ù: –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –æ—Ç–∫–ª—é—á–µ–Ω–∞")
        
        end_time = datetime.now()
        processing_time = (end_time - start_time).total_seconds()
        
        # –°–ë–û–†–ö–ê –†–ï–ó–£–õ–¨–¢–ê–¢–ê
        strict_result = {
            **result,
            'quality_metrics': quality_metrics,
            'strict_processing_info': {
                'transcription_method': transcription_method,
                'providers_used': list(self.providers.keys()),
                'ensemble_forced': force_ensemble,
                'preprocessing_forced': force_preprocessing,
                'quality_check_forced': force_quality_check,
                'processing_time_seconds': processing_time,
                'language': language,
                'strict_mode': True,
                'no_fallback': True
            }
        }
        
        # –í–´–í–û–î –†–ï–ó–£–õ–¨–¢–ê–¢–û–í
        print("\n" + "="*60)
        print("‚ö° –°–¢–†–û–ì–ò–ô –†–ï–ó–£–õ–¨–¢–ê–¢ –ü–†–û–§–ï–°–°–ò–û–ù–ê–õ–¨–ù–û–ô –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–ò")
        print("="*60)
        print(strict_result["text"])
        print("="*60)
        
        self._print_strict_quality_metrics(quality_metrics)
        
        print(f"\n‚è±Ô∏è  –í—Ä–µ–º—è –æ–±—Ä–∞–±–æ—Ç–∫–∏: {processing_time:.1f}—Å")
        print(f"üéØ –ú–µ—Ç–æ–¥: {transcription_method}")
        print(f"üî• –°—Ç—Ä–æ–≥–∏–π —Ä–µ–∂–∏–º: –ê–ö–¢–ò–í–ï–ù")
        
        return strict_result
    
    def _print_strict_quality_metrics(self, metrics: Dict):
        """–°—Ç—Ä–æ–≥–∏–π –≤—ã–≤–æ–¥ –º–µ—Ç—Ä–∏–∫ –∫–∞—á–µ—Å—Ç–≤–∞"""
        print("\nüìä –°–¢–†–û–ì–ò–ï –ú–ï–¢–†–ò–ö–ò –ö–ê–ß–ï–°–¢–í–ê:")
        print("‚îÄ" * 50)
        
        # –û—Å–Ω–æ–≤–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏
        key_metrics = [
            ('confidence_score', '–£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å', '%'),
            ('wer', 'Word Error Rate', '%'),
            ('cer', 'Character Error Rate', '%'),
            ('semantic_score', '–°–µ–º–∞–Ω—Ç–∏—á–µ—Å–∫–∞—è –æ—Ü–µ–Ω–∫–∞', '%'),
            ('ecommerce_score', 'E-commerce —Å–ø–µ—Ü–∏—Ñ–∏—á–Ω–æ—Å—Ç—å', '%')
        ]
        
        for key, name, unit in key_metrics:
            if key in metrics:
                value = metrics[key]
                if isinstance(value, float) and 0 <= value <= 1:
                    print(f"üéØ {name}: {value:.1%}")
                else:
                    print(f"üéØ {name}: {value}")
        
        print("‚îÄ" * 50)
    
    def get_strict_system_info(self) -> Dict:
        """–°–¢–†–û–ì–ê–Ø –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Å–∏—Å—Ç–µ–º–µ"""
        return {
            'strict_mode': True,
            'no_fallback': True,
            'device': self.device,
            'providers': {
                'active': list(self.providers.keys()),
                'count': len(self.providers)
            },
            'services': {
                'audio_processor': self.audio_processor is not None,
                'quality_assessor': self.quality_assessor is not None,
                'ensemble_service': self.ensemble_service is not None
            },
            'capabilities': {
                'ensemble_available': self.ensemble_service is not None,
                'quality_assessment': True,
                'audio_preprocessing': True,
                'russian_optimization': True,
                'ecommerce_specialization': True
            },
            'gpu_info': torch.cuda.get_device_properties(0).__dict__ if torch.cuda.is_available() else None
        }

# –°–¢–†–û–ì–ê–Ø –ò–ù–ò–¶–ò–ê–õ–ò–ó–ê–¶–ò–Ø
print("\n‚ö°‚ö°‚ö° –ù–ê–ß–ò–ù–ê–Æ –°–¢–†–û–ì–£–Æ –ò–ù–ò–¶–ò–ê–õ–ò–ó–ê–¶–ò–Æ ‚ö°‚ö°‚ö°")
strict_transcriber = StrictEnhancedTranscriber()
print("\nüéâüéâüéâ –°–¢–†–û–ì–ê–Ø –°–ò–°–¢–ï–ú–ê –ì–û–¢–û–í–ê –ö –†–ê–ë–û–¢–ï! üéâüéâüéâ")
````

````python
# üö® –°–¢–†–û–ì–ò–ï –î–ï–ú–û-–§–£–ù–ö–¶–ò–ò –ë–ï–ó FALLBACK
from google.colab import files
import json
from datetime import datetime
from pathlib import Path

def upload_and_transcribe_STRICT():
    """–°–¢–†–û–ì–ê–Ø –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω–∞—è —Ç—Ä–∞–Ω—Å–∫—Ä–∏–ø—Ü–∏—è - –ë–ï–ó –ö–û–ú–ü–†–û–ú–ò–°–°–û–í!"""
    print("‚ö°‚ö°‚ö° –°–¢–†–û–ì–ê–Ø –ü–†–û–§–ï–°–°–ò–û–ù–ê–õ–¨–ù–ê–Ø –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–Ø ‚ö°‚ö°‚ö°")
    print("üö® –†–ï–ñ–ò–ú: –ù–ò–ö–ê–ö–ò–• FALLBACK - –¢–û–õ–¨–ö–û –ü–û–õ–ù–ê–Ø –ú–û–©–ù–û–°–¢–¨!")
    print("="*60)
    
    try:
        # –°–¢–†–û–ì–ê–Ø –∑–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞
        print("üìÅ –ó–∞–≥—Ä—É–∑–∏—Ç–µ –∞—É–¥–∏–æ —Ñ–∞–π–ª –¥–ª—è –°–¢–†–û–ì–û–ì–û –∞–Ω–∞–ª–∏–∑–∞...")
        uploaded = files.upload()
        
        if not uploaded:
            raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –§–∞–π–ª –Ω–µ –±—ã–ª –∑–∞–≥—Ä—É–∂–µ–Ω!")
        
        filename = list(uploaded.keys())[0]
        filepath = Path(filename)
        
        # –ü–†–û–í–ï–†–ö–ê —Ñ–∞–π–ª–∞
        if not filepath.exists():
            raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω: {filename}")
        
        file_size = len(uploaded[filename]) / 1024 / 1024  # MB
        
        print(f"‚úÖ –§–ê–ô–õ –ü–†–ò–ù–Ø–¢:")
        print(f"   üìÑ –ò–º—è: {filename}")
        print(f"   üìä –†–∞–∑–º–µ—Ä: {file_size:.1f} MB")
        
        if file_size > 100:  # –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ 100MB
            raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –§–∞–π–ª —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–π ({file_size:.1f} MB > 100 MB)")
        
        # –°–¢–†–û–ì–ò–ï –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ - –í–°–ï –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–û
        strict_settings = {
            'language': 'ru',
            'force_ensemble': True,
            'force_preprocessing': True, 
            'force_quality_check': True
        }
        
        print(f"\n‚öôÔ∏è  –°–¢–†–û–ì–ò–ï –ù–ê–°–¢–†–û–ô–ö–ò (–ë–ï–ó –í–´–ë–û–†–ê):")
        print(f"   üåç –Ø–∑—ã–∫: {strict_settings['language']}")
        print(f"   ü§ñ –ê–Ω—Å–∞–º–±–ª—å: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–û")
        print(f"   üéµ –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–û")
        print(f"   üìä –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞: –ü–†–ò–ù–£–î–ò–¢–ï–õ–¨–ù–û")
        
        # –°–¢–†–û–ì–ê–Ø —Ç—Ä–∞–Ω—Å–∫—Ä–∏–ø—Ü–∏—è
        print("\nüöÄ –ù–ê–ß–ò–ù–ê–Æ –°–¢–†–û–ì–£–Æ –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–Æ...")
        print("‚ö†Ô∏è  –õ–Æ–ë–ê–Ø –û–®–ò–ë–ö–ê = –ü–û–õ–ù–ê–Ø –û–°–¢–ê–ù–û–í–ö–ê!")
        
        start_time = datetime.now()
        
        result = strict_transcriber.transcribe_professional_strict(
            str(filepath),
            **strict_settings
        )
        
        end_time = datetime.now()
        total_time = (end_time - start_time).total_seconds()
        
        print(f"\nüéâ –°–¢–†–û–ì–ê–Ø –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–Ø –ó–ê–í–ï–†–®–ï–ù–ê –£–°–ü–ï–®–ù–û!")
        print(f"‚è±Ô∏è  –û–±—â–µ–µ –≤—Ä–µ–º—è: {total_time:.1f}—Å")
        
        # –°–¢–†–û–ì–û–ï —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
        base_name = filepath.stem
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        
        files_to_create = {
            f"{base_name}_STRICT_RESULT_{timestamp}.json": result,
            f"{base_name}_STRICT_TEXT_{timestamp}.txt": result["text"],
        }
        
        # –°–æ–∑–¥–∞–Ω–∏–µ SRT –µ—Å–ª–∏ –µ—Å—Ç—å —Å–µ–≥–º–µ–Ω—Ç—ã
        if 'segments' in result and result['segments']:
            srt_content = create_srt_content(result['segments'])
            files_to_create[f"{base_name}_STRICT_SUBTITLES_{timestamp}.srt"] = srt_content
        
        # –°–æ–∑–¥–∞–Ω–∏–µ –æ—Ç—á–µ—Ç–∞ –æ –∫–∞—á–µ—Å—Ç–≤–µ
        quality_report = create_strict_quality_report(
            result.get('quality_metrics', {}),
            result.get('strict_processing_info', {}),
            total_time
        )
        files_to_create[f"{base_name}_STRICT_QUALITY_{timestamp}.txt"] = quality_report
        
        # –°–¢–†–û–ì–û–ï —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤
        print(f"\nüìÅ –°–û–ó–î–ê–Æ –°–¢–†–û–ì–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–´...")
        for filename, content in files_to_create.items():
            try:
                if filename.endswith('.json'):
                    with open(filename, 'w', encoding='utf-8') as f:
                        json.dump(content, f, ensure_ascii=False, indent=2, default=str)
                else:
                    with open(filename, 'w', encoding='utf-8') as f:
                        f.write(str(content))
                
                print(f"‚úÖ –°–æ–∑–¥–∞–Ω: {filename}")
            except Exception as e:
                raise RuntimeError(f"‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê —Å–æ–∑–¥–∞–Ω–∏—è {filename}: {e}")
        
        # –°–¢–†–û–ì–û–ï —Å–∫–∞—á–∏–≤–∞–Ω–∏–µ
        print("\nüì• –°–ö–ê–ß–ò–í–ê–Æ –í–°–ï –°–¢–†–û–ì–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–´...")
        for filename in files_to_create.keys():
            try:
                files.download(filename)
                print(f"üì• –°–∫–∞—á–∞–Ω: {filename}")
            except Exception as e:
                print(f"‚ö†Ô∏è  –ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ - –Ω–µ —É–¥–∞–ª–æ—Å—å —Å–∫–∞—á–∞—Ç—å {filename}: {e}")
        
        print("\nüéâüéâüéâ –°–¢–†–û–ì–ê–Ø –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–Ø –ó–ê–í–ï–†–®–ï–ù–ê! üéâüéâüéâ")
        return result
        
    except Exception as e:
        print(f"\nüí•üí•üí• –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –°–¢–†–û–ì–û–ô –°–ò–°–¢–ï–ú–´! üí•üí•üí•")
        print(f"‚ùå {e}")
        print("üö® –°–ò–°–¢–ï–ú–ê –û–°–¢–ê–ù–û–í–õ–ï–ù–ê - –ò–°–ü–†–ê–í–¨–¢–ï –û–®–ò–ë–ö–£!")
        raise

def batch_transcription_STRICT():
    """–°–¢–†–û–ì–ê–Ø –ø–∞–∫–µ—Ç–Ω–∞—è —Ç—Ä–∞–Ω—Å–∫—Ä–∏–ø—Ü–∏—è - –ë–ï–ó –ö–û–ú–ü–†–û–ú–ò–°–°–û–í!"""
    print("‚ö°‚ö°‚ö° –°–¢–†–û–ì–ê–Ø –ü–ê–ö–ï–¢–ù–ê–Ø –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–Ø ‚ö°‚ö°‚ö°")
    print("üö® –†–ï–ñ–ò–ú: –ö–ê–ñ–î–´–ô –§–ê–ô–õ –û–ë–†–ê–ë–ê–¢–´–í–ê–ï–¢–°–Ø –°–¢–†–û–ì–û!")
    print("="*60)
    
    try:
        # –°–¢–†–û–ì–ê–Ø –∑–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–æ–≤
        print("üìÅ –ó–∞–≥—Ä—É–∑–∏—Ç–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∞—É–¥–∏–æ —Ñ–∞–π–ª–æ–≤ –¥–ª—è –°–¢–†–û–ì–û–ô –æ–±—Ä–∞–±–æ—Ç–∫–∏...")
        uploaded = files.upload()
        
        if not uploaded:
            raise RuntimeError("‚ùå –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê: –§–∞–π–ª—ã –Ω–µ –±—ã–ª–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
        
        total_files = len(uploaded)
        print(f"‚úÖ –ü–†–ò–ù–Ø–¢–û –§–ê–ô–õ–û–í: {total_files}")
        
        # –°–¢–†–û–ì–ê–Ø –æ–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞–∂–¥–æ–≥–æ —Ñ–∞–π–ª–∞
        results = []
        errors = []
        
        for i, (filename, content) in enumerate(uploaded.items(), 1):
            print(f"\nüîÑ –°–¢–†–û–ì–û –û–ë–†–ê–ë–ê–¢–´–í–ê–Æ {i}/{total_files}: {filename}")
            
            try:
                # –°–¢–†–û–ì–ê–Ø –ø—Ä–æ–≤–µ—Ä–∫–∞ —Ñ–∞–π–ª–∞
                if not Path(filename).exists():
                    raise RuntimeError(f"–§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω: {filename}")
                
                # –°–¢–†–û–ì–ê–Ø —Ç—Ä–∞–Ω—Å–∫—Ä–∏–ø—Ü–∏—è
                result = strict_transcriber.transcribe_professional_strict(
                    filename,
                    language='ru',
                    force_ensemble=True,
                    force_preprocessing=True,
                    force_quality_check=True
                )
                
                results.append({
                    'filename': filename,
                    'success': True,
                    'result': result,
                    'error': None
                })
                
                print(f"‚úÖ {filename} - –°–¢–†–û–ì–û –ó–ê–í–ï–†–®–ï–ù")
                
            except Exception as e:
                error_info = {
                    'filename': filename,
                    'success': False,
                    'result': None,
                    'error': str(e)
                }
                results.append(error_info)
                errors.append(error_info)
                
                print(f"‚ùå {filename} - –°–¢–†–û–ì–ê–Ø –û–®–ò–ë–ö–ê: {e}")
        
        # –°–¢–†–û–ì–ò–ô –æ—Ç—á–µ—Ç
        successful = sum(1 for r in results if r['success'])
        failed = len(errors)
        
        print(f"\nüìä –°–¢–†–û–ì–ò–ô –ò–¢–û–ì–û–í–´–ô –û–¢–ß–ï–¢:")
        print(f"   üìÅ –í—Å–µ–≥–æ —Ñ–∞–π–ª–æ–≤: {total_files}")
        print(f"   ‚úÖ –£—Å–ø–µ—à–Ω–æ: {successful}")
        print(f"   ‚ùå –° –æ—à–∏–±–∫–∞–º–∏: {failed}")
        print(f"   üìä –£—Å–ø–µ—à–Ω–æ—Å—Ç—å: {successful/total_files:.1%}")
        
        # –°–¢–†–û–ì–û–ï —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Å–≤–æ–¥–Ω–æ–≥–æ –æ—Ç—á–µ—Ç–∞
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        batch_report_file = f"STRICT_BATCH_REPORT_{timestamp}.json"
        
        batch_summary = {
            'strict_mode': True,
            'timestamp': timestamp,
            'total_files': total_files,
            'successful': successful,
            'failed': failed,
            'success_rate': successful/total_files,
            'results': results
        }
        
        with open(batch_report_file, 'w', encoding='utf-8') as f:
            json.dump(batch_summary, f, ensure_ascii=False, indent=2, default=str)
        
        files.download(batch_report_file)
        
        if errors:
            print("\n‚ö†Ô∏è  –°–¢–†–û–ì–ò–ï –û–®–ò–ë–ö–ò –í –ü–ê–ö–ï–¢–ï:")
            for error_info in errors:
                print(f"   ‚ùå {error_info['filename']}: {error_info['error']}")
        
        print(f"\nüéâ –°–¢–†–û–ì–ê–Ø –ü–ê–ö–ï–¢–ù–ê–Ø –û–ë–†–ê–ë–û–¢–ö–ê –ó–ê–í–ï–†–®–ï–ù–ê!")
        return results
        
    except Exception as e:
        print(f"\nüí•üí•üí• –ö–†–ò–¢–ò–ß–ï–°–ö–ê–Ø –û–®–ò–ë–ö–ê –ü–ê–ö–ï–¢–ù–û–ô –°–ò–°–¢–ï–ú–´! üí•üí•üí•")
        print(f"‚ùå {e}")
        raise

def create_srt_content(segments):
    """–°–æ–∑–¥–∞–Ω–∏–µ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ SRT —Ñ–∞–π–ª–∞"""
    srt_content = []
    for i, segment in enumerate(segments, 1):
        start_time = seconds_to_srt_time(segment['start'])
        end_time = seconds_to_srt_time(segment['end'])
        srt_content.append(f"{i}")
        srt_content.append(f"{start_time} --> {end_time}")
        srt_content.append(segment['text'].strip())
        srt_content.append("")  # –ü—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞
    return "\n".join(srt_content)

def seconds_to_srt_time(seconds):
    """–ö–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏—è —Å–µ–∫—É–Ω–¥ –≤ SRT –≤—Ä–µ–º—è"""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    millisecs = int((seconds % 1) * 1000)
    return f"{hours:02d}:{minutes:02d}:{secs:02d},{millisecs:03d}"

def create_strict_quality_report(quality_metrics, processing_info, total_time):
    """–°–æ–∑–¥–∞–Ω–∏–µ —Å—Ç—Ä–æ–≥–æ–≥–æ –æ—Ç—á–µ—Ç–∞ –æ –∫–∞—á–µ—Å—Ç–≤–µ"""
    report = []
    report.append("‚ö° –°–¢–†–û–ì–ò–ô –û–¢–ß–ï–¢ –û –ö–ê–ß–ï–°–¢–í–ï –¢–†–ê–ù–°–ö–†–ò–ü–¶–ò–ò")
    report.append("=" * 60)
    report.append("")
    
    # –û—Å–Ω–æ–≤–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è
    report.append(f"üìÖ –î–∞—Ç–∞ –∏ –≤—Ä–µ–º—è: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    report.append(f"‚è±Ô∏è  –í—Ä–µ–º—è –æ–±—Ä–∞–±–æ—Ç–∫–∏: {total_time:.1f} —Å–µ–∫—É–Ω–¥")
    report.append(f"üö® –†–µ–∂–∏–º: –°–¢–†–û–ì–ò–ô (–ë–ï–ó FALLBACK)")
    report.append("")
    
    # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –æ–±—Ä–∞–±–æ—Ç–∫–µ
    if processing_info:
        report.append("üîß –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –û–ë–†–ê–ë–û–¢–ö–ï:")
        report.append("-" * 40)
        for key, value in processing_info.items():
            if isinstance(value, bool):
                status = "‚úÖ –î–ê" if value else "‚ùå –ù–ï–¢"
                report.append(f"   {key}: {status}")
            elif isinstance(value, list):
                report.append(f"   {key}: {', '.join(value)}")
            else:
                report.append(f"   {key}: {value}")
        report.append("")
    
    # –ú–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞
    if quality_metrics:
        report.append("üìä –°–¢–†–û–ì–ò–ï –ú–ï–¢–†–ò–ö–ò –ö–ê–ß–ï–°–¢–í–ê:")
        report.append("-" * 40)
        
        for key, value in quality_metrics.items():
            if isinstance(value, float) and 0 <= value <= 1:
                report.append(f"   üéØ {key}: {value:.1%}")
            else:
                report.append(f"   üéØ {key}: {value}")
        report.append("")
    
    report.append("üéâ –°–¢–†–û–ì–ò–ô –û–¢–ß–ï–¢ –ó–ê–í–ï–†–®–ï–ù")
    report.append("=" * 60)
    
    return "\n".join(report)

print("üö®üö®üö® –°–¢–†–û–ì–ò–ï –î–ï–ú–û-–§–£–ù–ö–¶–ò–ò –°–û–ó–î–ê–ù–´! üö®üö®üö®")
print("\n‚ö° –î–û–°–¢–£–ü–ù–´–ï –°–¢–†–û–ì–ò–ï –§–£–ù–ö–¶–ò–ò:")
print("   üéØ upload_and_transcribe_STRICT() - –°–¢–†–û–ì–ê–Ø –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω–∞—è —Ç—Ä–∞–Ω—Å–∫—Ä–∏–ø—Ü–∏—è")
print("   üì¶ batch_transcription_STRICT() - –°–¢–†–û–ì–ê–Ø –ø–∞–∫–µ—Ç–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞")
print("\nüö® –í–ù–ò–ú–ê–ù–ò–ï: –ë–ï–ó FALLBACK - –¢–û–õ–¨–ö–û –ü–û–õ–ù–ê–Ø –°–ò–°–¢–ï–ú–ê!")
````
